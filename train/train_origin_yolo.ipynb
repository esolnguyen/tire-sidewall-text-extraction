```python
import torch
from ultralytics import YOLO
```

```python
model = YOLO("/content/drive/My Drive/tire_extraction/yolo_experiment1/outputs/Original_YOLO_E1002/weights/last.pt")
```

```python
model.train(
    task="segment",
    data="coco.yaml",
    epochs=300,
    project='/content/drive/My Drive/tire_extraction/yolo_experiment1/outputs',
    resume=True,
)
```

```python
for p in model.model.parameters():
    p.requires_grad = True

model.train(
    task="segment",
    data="coco.yaml",
    epochs=100,
    batch=16,
    rect=True,
    device=0,
    workers=8,
    optimizer="AdamW",
    lr0=0.001,
    lrf=0.01,
    cos_lr=True,
    patience=30,
    amp=True,
    project='/content/drive/My Drive/tire_extraction/yolo_experiment1/outputs',
    resume=True,
)
```
